```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

szery = []

all_deaths = pd.read_csv(Path('dane/smierci.csv'))
all_deaths

for i in range(0,13 ):
#    print( f'{i}  {(all_deaths.iloc[i,0]+all_deaths.iloc[i+1,0])/2}   {(all_deaths.iloc[i+1,0]-all_deaths.iloc[i,0])} ')
    szery.append(((all_deaths.iloc[i,0]+all_deaths.iloc[i+1,0])/2, all_deaths.iloc[i+1,0]-all_deaths.iloc[i,0], all_deaths.iloc[i,1]-300_000))

#plt.bar()
print(szery)

do_rysowania = pd.DataFrame(szery)
plt.bar(x= do_rysowania[0], height=do_rysowania[2], width = do_rysowania[1])
plt.xticks(ticks = all_deaths['rok'], labels = all_deaths['rok'], rotation = 45)
plt.yticks(ticks = [x * 10_000 for x in range(0,20,2)], labels = [x * 10_000 + 300_000 for x in range(0,20,2)], rotation = 45)
plt.grid('both')
plt.show()
```
